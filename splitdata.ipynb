```python
#Spliting data uisng python and clean and manege data

import pandas as pd

# Sample data in the format similar to the provided image
data = {
    'ADDR': ['45 rd, kenith street, btm, bangalore 500038', '45 rd, street, hitech, Hd 500038'],
    'DT': ['03-04-2022', '02-04-2022'],
    'NAME': ['abc', 'dec'],
    'time': ['8:30:21', '5:30:21']
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Splitting the address into separate components
df[['House add', 'street', 'city', 'state_zip']] = df['ADDR'].str.split(',', expand=True)

# Further splitting the 'state_zip' column to separate state and zip
df[['state', 'zip']] = df['state_zip'].str.rsplit(' ',n=1,expand=True)
# Dropping the original columns that have been split
df.drop(columns=['ADDR', 'state_zip'], inplace=True)

# Splitting the date into day, month, and year
df[['day', 'month', 'year']] = df['DT'].str.split('-', expand=True)

# Dropping the original date column
df.drop(columns=['DT'], inplace=True)

# Splitting the time into hours, minutes, and seconds
df[['hr', 'min', 'sec']] = df['time'].str.split(':', expand=True)

# Dropping the original time column
df.drop(columns=['time'], inplace=True)

# Displaying the cleaned DataFrame
print(df)
```

```python
import re
def clean_char(text):
    if pd.isnull(text):
        return text
    return re.sub(r"[^A-Za-z0-9]+", "", str(text))

df["Name"]=df["Name"].apply(clean_char)   
df["Domain"]=df["Domain"].apply(clean_char)
df["Salary"]=df["Salary"].apply(clean_char)
df
```

```python

df['Age'] = df['Age'].str.extract(r'(\d+)').astype(float)
df["Exp"]=df["Exp"].str.extract(r'(\d+)').astype(float)
```

```python
df["Location"].fillna("Unknown", inplace=True)
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Exp"].fillna(df["Exp"].mean(), inplace=True)
df
```

```python
df.to_excel('cleaned_data.xlsx', index=False)

print("\nCleaned data saved to 'cleaned_data.xlsx'")
```
